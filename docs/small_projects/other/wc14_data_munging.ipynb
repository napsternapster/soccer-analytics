```python
onename_country_matches = []
twoname_country_matches = []
twoname_country_matches2 = []

for i in matches_semiclean:
    i = [j for j in i if j!='vs']
    if (len(i) == 3):        
        onename_country_matches.append(i)
    else:
        twoname_country_matches.append(i)
        
for i in twoname_country_matches:
    i = [j for j in i if j!='herz']
    twoname_country_matches2.append(i)
            
for i in twoname_country_matches2:
    if('bosnia' in i):
        onename_country_matches.append(i)
        twoname_country_matches2 = [j for j in twoname_country_matches2 if j != i]

twoname_country_matches2a = []
twoname_country_matches2b = []

for i in twoname_country_matches2:
    if(i[0]=='south' or i[0]== 'costa' or i[0]=='ivory'):
        twoname_country_matches2a.append(i)
    else:
        twoname_country_matches2b.append(i)

for i in twoname_country_matches2a:
    i[1] = i[0] + ' '+i[1]
    i = i[1:]
    onename_country_matches.append(i)
    
for i in twoname_country_matches2b:
    i[1] = i[1] +' '+i[2]
    i[2] = i[3]
    i = i[:3]
    onename_country_matches.append(i)
```

It turns out that the dataframe that I created before weren't as immediately useful as I thought they would have been. I was stuck in No man's land between being overly general, and overly specific, with the latter being my goal to analyze Brazil versus Germany. Luckily, someone suggested that I create a dataframe that would be conducive to building a model to try to predict the score of Brazil v Germany based on stats only. 

The dataframe is composed of all 64 games from the World Cup, and the difference in frequency for each statistics between the two competitors. 

Here's how it was constructed. (Brazil v Germany was added first.)

```python
brazil_v_germany_id = '731827'

df_data2[brazil_v_germany_id][0]['team_stats'].keys()
```

```python
bra_ger_tstats = pd.DataFrame.from_dict(df_data2[brazil_v_germany_id][0]['team_stats'], orient='index')
bra_ger_tstats = bra_ger_tstats.fillna(0)

diff = bra_ger_tstats.ix[0]-bra_ger_tstats.ix[1]
diff['team1']=bra_ger_tstats.index[0]
diff['team2']=bra_ger_tstats.index[1]

temp = pd.DataFrame(diff)

if temp.ix['goals'][0]==0.0:
    temp['result']=0 #draw
    
elif temp.ix['goals'][0]>0:
    temp['result']=1 #win
    
elif temp.ix['goals'][0]<0.0:
    temp['result']=-1 #loss
```

Now I add the stats from the other 63 games.

```python
#loop through df_matches match ids

count = 0
for match in df_matches.match_id:
#for each one create a series

    df_temp = pd.DataFrame.from_dict(df_data2[match][0]['team_stats'], orient='index')
    df_temp = df_temp.fillna(0)
    diff = df_temp.ix[0]-df_temp.ix[1]
    diff['team1']=df_temp.index[0]
    diff['team2']=df_temp.index[1]
    temp = pd.DataFrame(diff).transpose()
    
    try:
        if temp['goals'][0]==0.0:
            temp['result']=0
        elif temp['goals'][0]>0:
            temp['result']=1
        elif temp['goals'][0]<0.0:
            temp['result']=-1
    except:
        temp['result']=0
    
    #concat to the diff dataframe and repeat
    if count==0:
        df_all = temp
    else:
        df_all = pd.concat([df_all,temp])
    count+=1
    
print(df_all.head())
```
