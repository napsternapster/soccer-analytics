```python
import pandas as pd
data = pd.read_csv('mls_results.csv')
data['score'] = data['selection1_score'].map(lambda x: x if 'n.P' in x else 0)
data_hockey = data.ix[data.score!=0,:]
data_hockey = data_hockey.ix[:, [0,1,5]]
```

```python
data_hockey['home_score'] = data_hockey['score'].map(lambda x: x[:-5].split(':')[0])
data_hockey['away_score'] = data_hockey['score'].map(lambda x: x[:-5].split(':')[1])
data_hockey['score'] = data_hockey['score'].map(lambda x: x[:-5])
```
